```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = "https://www.billboard.com/charts/hot-100"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup.get_text()
```

I found two relevant HTML elements: 
   - title-of-a-story 

and 

- o-chart-results-list__item

Both require significant cleaning, so I'll just pick the second one.


```python
mylist = []
```

```python
for i in range(len(soup.select(".o-chart-results-list__item"))):
    if re.search("[A-z]", soup.select(".o-chart-results-list__item")[i].get_text()):
        mylist.append(soup.select(".o-chart-results-list__item")[i].get_text())
```

```python
mylist2 = []
for item in mylist: 
    mylist2.append(item.strip())
```

```python
mylist3 = []
for item in mylist2:
    mylist3.append(item.split("\t"))
    
```

```python
mylist3
```

Wow, this is still very ugly. But at least I now have titles as the first element in the list, and artist as the last. (And then I'll still need to get rid of the "NEW" columns)

```python
mylist4 = []
```

```python
for item in mylist3:
    mylist4.append([item[0],item[-1]])
```

```python
mylist4
```

```python
mylist5 = []
```

```python
for item in mylist4: 
    if "NEW" not in item:
        if "RE-\nENTRY" not in item:
            mylist5.append(item)
```

```python
mylist5
```

Beautiful.

```python
df = pd.DataFrame(mylist5, columns = ['Title', 'Artist'])
```
