## Game plan

What my function will need to do:
- request page
- parse in beautiful soup
- get list of titles
- get list of artists
- in list of titles ("td:nth-child(3)"), get rid of any entries that are not in double quotes
- in list of artists (("td:nth-child(4)"), get rid of any entries that consist only of numbers


```python
def scrappucino(list_of_webpages):
    artists = []
    titles = []
    for page in list_of_webpages:
        webpage = requests.get(page)
        soup = BeautifulSoup(webpage.content, 'html.parser')
        
        # Get titles (getting rid of anything without double quotes first, then eliminating double quotes)
        for item in soup.select("td:nth-child(3)"):
            if '"' in item.get_text(): 
                titles.append(item.get_text().replace('"',''))
        # Get artists (getting rid of mere numbers)
        for item in soup.select("td:nth-child(4)"):
            if not item.get_text().isnumeric():
                artists.append(item.get_text())
    print(len(artists))
    print(len(titles))
    return
```

## Function, take 2

What I'll do to solve the problem is to remove empty strings within the function, as a separate step. 

```python
def scrappucino(list_of_pages):
    artists = []
    titles = []
    for page in list_of_pages:
        webpage = requests.get(page)
        soup = BeautifulSoup(webpage.content, 'html.parser')

    # Get titles (getting rid of anything without double quotes first, then eliminating double quotes)
        for item in soup.select("td:nth-child(3)"):
            if '"' in item.get_text(): 
                titles.append(item.get_text().replace('"','').strip())

        # Get artists (getting rid of mere numbers)
        for item in soup.select("td:nth-child(4)"):
            if item.get_text().strip().isnumeric() == False:
                artists.append(item.get_text().strip())

    if '' in artists: 
        artists.remove('')
    
    data = pd.DataFrame()
    data['Title'] = titles
    data['Artist'] = artists
    return data
```
