## Game plan

What my function will need to do:
- request page
- parse in beautiful soup
- get list of titles
- get list of artists
- in list of titles ("td:nth-child(3)"), get rid of any entries that are not in double quotes
- in list of artists (("td:nth-child(4)"), get rid of any entries that consist only of numbers


```python
def scrappucino(list_of_webpages):
    artists = []
    titles = []
    for page in list_of_webpages:
        webpage = requests.get(page)
        soup = BeautifulSoup(webpage.content, 'html.parser')
        
        # Get titles (getting rid of anything without double quotes first, then eliminating double quotes)
        for item in soup.select("td:nth-child(3)"):
            if '"' in item.get_text(): 
                titles.append(item.get_text().replace('"',''))
        # Get artists (getting rid of mere numbers)
        for item in soup.select("td:nth-child(4)"):
            if not item.get_text().isnumeric():
                artists.append(item.get_text())
    print(len(artists))
    print(len(titles))
    return
```

## Function, take 2

What I'll do to solve the problem is to remove empty strings within the function, as a separate step. 

```python
def scrappucino(list_of_pages):
    artists = []
    titles = []
    for page in list_of_pages:
        webpage = requests.get(page)
        soup = BeautifulSoup(webpage.content, 'html.parser')

    # Get titles (getting rid of anything without double quotes first, then eliminating double quotes)
        for item in soup.select("td:nth-child(3)"):
            if '"' in item.get_text(): 
                titles.append(item.get_text().replace('"','').strip())

        # Get artists (getting rid of mere numbers)
        for item in soup.select("td:nth-child(4)"):
            if item.get_text().strip().isnumeric() == False:
                artists.append(item.get_text().strip())

    if '' in artists: 
        artists.remove('')
    
    data = pd.DataFrame()
    data['Title'] = titles
    data['Artist'] = artists
    return data
```

# Now to automate that into a single reusable function

```python
def convert_playlist(playlist_id):
    # First, do what get_playlist_tracks did:
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results["items"]
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3000)/1000)

    print("Done with retrieving playlist. Moving on.")    
    # Then get the relevant info: 
    titles = []
    artists = []
    uris = []

    for entry in tracks: 
        titles.append(entry['track']['name'])
        artists.append(entry['track']['artists'][0]['name']) # This is the lazy approach - using only the primary artist.
        uris.append(entry['track']['uri'])
        
    # Set up the dataframe
    df_dict = {'titles': titles, 'artists': artists, 'uri': uris} 
    songs_df = pd.DataFrame(data=df_dict)
    
    # Retrieve list of audio features
    print("Starting on retrieving audio features...")
    featureslist = []
    deletionslist = []
    for uri in uris: #This is where the list of uris comes into play
        try:
            featureslist.append(sp.audio_features(uri))
            sleep(randint(1,4000)/2000)
        except:
            print(f"Skipping {uri}")
            deletionslist.append(uri)
            
    print("Done retrieving audio features")

    # Clean up the original dataframe.
    for item in deletionslist: 
        songs_df = songs_df[songs_df[uri] != item].copy()
        # This is necessary in case the server times out during the try/except loop above (which it does - often).
        # If I don't clean it up I will get more song titles than audio features.
        # Which means a bunch of NaNs at the bottom of the dataframe, and a mismatch between titles and features
    
    # Normalize the audio features (which immediately also generates a dataframe)
    features_df = pd.DataFrame()

    for feature in featureslist: 
        features_df = pd.concat([features_df,pd.json_normalize(feature)], axis = 0).copy()
    # Reset the index
    features_df = features_df.reset_index().copy()
    
    # And concatenate the dataframe with songs and the dataframe with audio features
    songs_df = pd.concat([songs_df,features_df], axis = 1).copy()
    
    return songs_df
    
```
